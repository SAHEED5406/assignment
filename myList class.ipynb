```python
import logging as log

log.basicConfig(filename="myList.log", filemode='w', level=log.DEBUG,
                format="%(asctime)s :  %(levelname)s : %(message)s")


class myList:

    def __init__(self, *args):
        if len(args) == 0:
            self.mylist = list(args)
            log.info("created myList object with zero elements")
        else:
            self.mylist = list(args)
            log.info(f"created myList object with {args}")

    def lenOfmyList(self):
        """THIS METHOD RETURNS NUMBER OF ELEMENTS PRESENT IN myList OBJECT"""
        log.info(f"Number of elements present in myList is{len(self.mylist)}")
        return len(self.mylist)

    def addTomyList(self, *args):
        """ THIS METHOD ADDS ELEMENTS TO myList OBJECT """
        for i in args:
            self.mylist.append(i)
        log.info(f"added {args} to myList ")

    def extendmyList(self, *args):
        """THIS METHOD ADDS ELEMENTS TO myList OBJECT AS A SEPERATE SINGLE ENTITY"""
        try:
            for i in args:
                self.mylist.extend(i)
                log.info(f"added {args} to myList ")
        except TypeError as e:
            log.exception(e)
            log.error("Entered int type ,converted to list type and added to myList ")
            return self.mylist.extend([args[0]])
        else:
            return self.mylist

    def insertInList(self, index, element):
        """ THIS METHOD INSERTS ELEMENTS TO myList OBJECT AT A SPECIFIC INDEX, IF INDEX IS OUT OF RANGE IT WOULD BE
        ADDED AT THE END """
        self.mylist.insert(index, element)
        if index <= len(self.mylist):
            log.info(f"added {element} at  index {index} ")
        else:
            log.info(f"{index} index is out of range , added {element} at  index {len(self.mylist) - 1} ")
        return self.mylist

    def clearmyList(self, *args):
        """ THIS METHOD CLEARS THE ELEMENTS PRESENT IN myList OBJECT """
        if len(args) == 0:
            log.info(f"removed {self.mylist} from myList ")
            return self.mylist.clear()
        else:
            log.error("clearmyList method needs no arguments , ignored the input given and cleared myList")
            return self.mylist.clear()

    def countInList(self, a):
        """THIS METHOD RETURNS NUMBER OF OCCURENCE OF AN ELEMENT IN myList OBJECT """
        log.info(f"occurance of {a} is {self.mylist.count(a)}")
        return self.mylist.count(a)

    def reversemyList(self):
        """THIS METHOD RETURNS THE REVERSED myList OBJECT """
        self.mylist.reverse()
        log.info(f"reversed myList is {self.mylist}")
        return self.mylist

    def removeFromList(self, *elements):
        """ THIS METHOD REMOVES THE GIVEN ELEMENTS FROM myList  """
        if len(elements) == 0:
            log.error("No Elements were given to remove , myList stays as it is  ")
        else:
            for element in elements:
                try:
                    self.mylist.remove(element)
                    log.info(f"removed {element} from myList")
                except ValueError as msg:
                    log.error(f" {element} is not available in myList , so this element is skipped ")
                    log.exception(msg)
                    continue
        return self.mylist

    def stopLog(self):
        """ THIS METHOD IS FOR STOPPING THE LOGGING Of myList """
        log.info(f"myList at the end of logging is {self.mylist}")
        log.info("END OF LOGGING")
        log.shutdown()

    def __str__(self):
        return "This is myList class which is equivalent list class"

```
